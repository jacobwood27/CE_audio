```julia
using Random
using DelimitedFiles
```

```julia
dat = readdlm("dat.txt", '|', String, '\n')
```

```julia
for row in eachrow(dat[107:end,:])
    title = row[1]
    url = row[2]

    cmd = `youtube-dl -o audio_dl/$(title).mp3 --extract-audio --audio-format mp3 $url`
    run(cmd)
end
```

```julia
d = readdir("audio_dl")
s = d[30]
cmd = `ffmpeg -i audio_dl/$s -f mp3 -vn audio_only/$(s[1:end-5] * ".mp3")`
run(cmd)
```

```julia
open("feed.xml","w") do io
  println(io,"<rss version=\"2.0\">")
  println(io,"<channel>")
  println(io,"<title>CE Lectures</title>")
  println(io,"<language>en-us</language>")
  println(io,"<description>Private podcast containing the lectures in the 2022 CE program.</description>")

  for row in eachrow(dat)
      title = row[1]
      uid = row[3]
      println(io,"<item><title>$title</title><enclosure url=\"https://drive.google.com/uc?export=download&id=$(uid[2:end])\" type=\"audio/mpeg\" /><guid>$(randstring(20))</guid><pubDate>Tue, 5 Jul 2022 09:00:00 +0000</pubDate></item>")
  end

  println(io,"</channel>")
  println(io,"</rss>")
end
```
